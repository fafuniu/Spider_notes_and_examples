# 如何判断是否是动态网页

动态网页随着网页的翻动会不断的在网页上添加新的内容。
两种方法：
1. selenium
2. 分析访问请求
   分析get请求，（XHR，或其他，寻找json文件，找到其对应的URL）（有些信息鼠标悬浮在一些图片上之后其会弹出窗口信息，这些信息也是动态产生的）

```python
import requests 
import re
import json
```

```python


def get_html(url):
    headers={"user-agent":'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:70.0) Gecko/20100101 Firefox/70.0'}
    response=requests.get(url,headers=headers)
    text=response.content.decode("utf-8")
    return text
def phase_html(text,num):
    #get rid of the wrong start
    text=re.findall("{.*}",text)[0]
    jtext=json.loads(text)
    for j in jtext["data"]:
        print(num, j["title"].strip())
        num=num+1
    return num

def main(): 
    num=0
    for i in range(10):
        url="http://staticize.mop.com/subject/v1/web/latest/dzh/post?callback=jQuery18306804594090706239_1573775865824&pageNo={}&pageSize=10&_=1573775867657".format(i)
        text=get_html(url)
        num=phase_html(text,num)

main()
```
