```python
from lxml import etree
import requests 
import re
```

```python
def get_html(url):
    headers={"user-agent":'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:70.0) Gecko/20100101 Firefox/70.0'}
    response=requests.get(url,headers=headers)
    content=response.content
    return content
def phase_html(html,num):
    x=etree.HTML(html)    
    #/html/body/div[3]/div[1]/div/div[1]/ol/li")
    zwmc=x.xpath("/html/body/div[3]/div[1]/div/div[1]/ol/li")
    for i in zwmc:
        film__name=i.xpath("div/div[2]/div[1]/a/span/text()")[0]
        director=i.xpath("div/div[2]/div[2]/p[1]/text()")[0]
        year=i.xpath("div/div[2]/div[2]/p[1]/text()")[1]
        print(num,film__name,director.strip(),year.strip())
        num=num+1
    return num
def main():
    num=0
    for i in range(10):
        url="https://movie.douban.com/top250?start={}&filter=".format(25*i)
        text=get_html(url).decode("utf-8")
        num=phase_html(text,num)

main()
```
