```python
import requests 
import re

def get_html(url):
    headers={"user-agent":'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:70.0) Gecko/20100101 Firefox/70.0'}
    response=requests.get(url,headers=headers)
    content=response.content
    return content
def phase_html(html):
    pattern= re.compile(r'src="(https://imgsa.baidu.com/forum.+?jpg)"')
    result=re.findall(pattern,html)
    return result
def save_pic(content,num):
    with open("fig/%s.jpg"%num,"wb") as f:
        f.write(content)
def main():
    num=0
    for i in range(6):
        url="https://tieba.baidu.com/p/5815297430?pn={}".format(i+1)
        text=get_html(url).decode("utf-8")
        url_list=phase_html(text)
        for one in url_list:
            print(num)
            try:
                content=get_html(one)
                save_pic(content,num)
                num+=1
            except:
                print("Error")

main()
```

# Example 2 猫眼电影ｔｏｐ１００


```python
import requests 
import re

def get_html(url):
    headers={"user-agent":'Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:70.0) Gecko/20100101 Firefox/70.0'}
    response=requests.get(url,headers=headers)
    content=response.content
    return content
def phase_html(html):
    pattern= re.compile('<dd>.+?index-(\d{1,3}).+?title="(\w+)".+?data-src="(.+?)".+?star">\s+(.+?)\s+</p>.+?releasetime">(.+?)</p>.+?class="integer">(\d\.)</i>.+?fraction">(\d).+?</dd>',re.S)
    result=re.findall(pattern,html)
    return result
def main():
    num=0
    for i in range(10):
        url="https://maoyan.com/board/4?offset={}".format(10*i)
        text=get_html(url).decode("utf-8")
        url_list=phase_html(text)
        for one in url_list:
            print(one[0],one[1])

main()
```
